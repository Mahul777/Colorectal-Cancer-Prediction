## Outliers

```python
# Step 1: Boxplots
# List of numerical columns to check for outliers
numerical_columns = ['Age', 'Tumor_Size_mm', 'Healthcare_Costs', 'Incidence_Rate_per_100K', 'Mortality_Rate_per_100K']

# Set up the plotting environment
plt.figure(figsize=(12, 8))

# Loop through the numerical columns and plot a boxplot for each
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=data[col].dropna())
    plt.title(f'Boxplot of {col}')

plt.tight_layout()
plt.show()
```

```python
# Step 2: Z-Score Method
from scipy.stats import zscore

# Calculate Z-scores
z_scores = np.abs(zscore(data[numerical_columns]))

# Flag data points with Z-score > 3 as outliers
outliers_z = (z_scores > 3)

# Display rows with outliers (True values indicate outliers)
outlier_data_z = data[numerical_columns][outliers_z.any(axis=1)]
outlier_data_z
''' 
| Age | Tumor_Size_mm | Healthcare_Costs | Incidence_Rate_per_100K | Mortality_Rate_per_100K |
| --- | ------------- | ---------------- | ----------------------- | ----------------------- |
| 30  | 52            | 34977            | 26                      | 8                       |
| 33  | 23            | 104925           | 22                      | 8                       |
| 31  | 16            | 117243           | 23                      | 7                       |
| 30  | 79            | 26696            | 31                      | 27                      |
| 32  | 25            | 64859            | 35                      | 24                      |
| ... | ...           | ...              | ...                     | ...                     |
| 31  | 33            | 37093            | 17                      | 11                      |
| 32  | 55            | 38281            | 45                      | 12                      |
| 30  | 9             | 75538            | 54                      | 17                      |
| 33  | 14            | 43834            | 52                      | 29                      |
| 31  | 76            | 118567           | 53                      | 18                      |

'''
```

```python
# Step 3: IQR Method
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = data[numerical_columns].quantile(0.25)
Q3 = data[numerical_columns].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define the lower and upper bounds for detecting outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Flag data points outside the lower and upper bounds as outliers
outliers_iqr = ((data[numerical_columns] < lower_bound) | (data[numerical_columns] > upper_bound))

# Display rows with outliers (True values indicate outliers)
outlier_data_iqr = data[numerical_columns][outliers_iqr.any(axis=1)]
outlier_data_iqr
''' 
| Age | Tumor_Size_mm | Healthcare_Costs | Incidence_Rate_per_100K | Mortality_Rate_per_100K |
| --- | ------------- | ---------------- | ----------------------- | ----------------------- |

'''
```

```python
# Target and Features Variable Selection
X = data.drop(columns=['Survival_Prediction'])
Y = data["Survival_Prediction"]

# Split the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# selecting best features to train our model
#chi-square test
from sklearn.feature_selection import chi2, SelectKBest

### CHI-SQUARe-TEST
X_cat = X_train.select_dtypes(include=['int64' , 'float64'])
chi2_selector = SelectKBest(score_func=chi2 , k="all")
chi2_selector.fit(X_cat,Y_train)
```

```python
chi2_scores = pd.DataFrame({
    'Feature' : X_cat.columns,
    "Chi2 Score" : chi2_selector.scores_
}).sort_values(by='Chi2 Score' , ascending=False)

chi2_scores
''' 
| Feature                    | Chi2 Score  |
| -------------------------- | ----------- |
| Patient_ID                 | 4648.993106 |
| Healthcare_Costs           | 2097.430133 |
| Tumor_Size_mm              | 7.971461    |
| Country_ Italy             | 6.957019    |
| Treatment_Type             | 4.987144    |
| Country_ France            | 4.266801    |
| Diabetes                   | 2.103967    |
| Mortality_Rate_per_100K    | 1.963077    |
| Country_ Pakistan          | 1.729603    |
| Country_ Canada            | 1.488030    |
| Insurance_Status           | 1.464241    |
| Cancer_Stage               | 1.281849    |
| Screening_History          | 1.159519    |
| Country_ India             | 0.934652    |
| Healthcare_Access          | 0.713346    |
| Age                        | 0.679739    |
| Survival_5_years           | 0.545494    |
| Country_ China             | 0.475911    |
| Diet_Risk                  | 0.445386    |
| Economic_Classification    | 0.381835    |
| Urban_or_Rural             | 0.355363    |
| Smoking_History            | 0.280786    |
| Country_ Japan             | 0.252218    |
| Country_ Germany           | 0.234813    |
| Alcohol_Consumption        | 0.209884    |
| Incidence_Rate_per_100K    | 0.208835    |
| Country_ USA               | 0.202055    |
| Obesity_BMI                | 0.151591    |
| Country_ South Africa      | 0.125858    |
| Family_History             | 0.095201    |
| Early_Detection            | 0.063440    |
| Mortality                  | 0.056049    |
| Genetic_Mutation           | 0.049299    |
| Physical_Activity          | 0.048910    |
| Country_ Nigeria           | 0.046787    |
| Country_ New Zealand       | 0.028651    |
| Country_ UK                | 0.022787    |
| Country_ Brazil            | 0.015730    |
| Gender                     | 0.011476    |
| Inflammatory_Bowel_Disease | 0.001772    |
| Country_ South Korea       | 0.001390    |

'''

# top 5 features
top_feaure = chi2_scores.head(8)['Feature'].tolist()
top_feaure
''' 
['Patient_ID',
 'Healthcare_Costs',
 'Tumor_Size_mm',
 'Country_ Italy       ',
 'Treatment_Type',
 'Country_ France      ',
 'Diabetes',
 'Mortality_Rate_per_100K']
'''

X = X[top_feaure]
X.shape # (167497, 8)
Y.shape #(167497,)

# train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

# Now scaling the data 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Model Training
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)
gb.fit(X_train, Y_train)

y_pred = gb.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test, y_pred)
print("Accuracy:", accuracy) # Accuracy: 0.598955223880597

# classification report
from sklearn.metrics import classification_report
report = classification_report(Y_test, y_pred)
print(report)
''' 
precision    recall  f1-score   support

           0       0.45      0.01      0.01     13412
           1       0.60      0.99      0.75     20088

    accuracy                           0.60     33500
   macro avg       0.52      0.50      0.38     33500
weighted avg       0.54      0.60      0.45     33500
'''
#  ROC AUC Score 
# This works specifically for Binary Classification (Yes/No).
# 	 Probability Prediction: To calculate this, we use predict_proba() instead of standard predict().

y_proba = gb.predict_proba(X_test)[:, 1]
from sklearn.metrics import roc_auc_score
roc_auc = roc_auc_score(Y_test, y_proba)
print("ROC AUC:", roc_auc) # ROC AUC: 0.49772988301221127
# 	 Interpretation: This is slightly better than a random coin toss. Again, this low score is a reflection of the synthetic data quality, not necessarily a flaw in your code.

```
